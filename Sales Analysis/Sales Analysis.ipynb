## 4th Insight: What products are most often sold together?

```python
df['Product'].unique()
```

### Checking the duplicate item in Order Id

```python
df=df[df["Order ID"].duplicated(keep=False)]
df
```

### Adding the Grouped Column which have a duplicated item of Order ID and Product

```python
df["Grouped"]=df.groupby("Order ID")["Product"].transform(lambda x: ','.join(x))
df
```

### Dropping the duplicate item that is common with Order ID

```python
df2=df[["Order ID","Grouped"]].drop_duplicates()
df2
```

### counting-unique-pairs-of-numbers-into-a-python-dictionary

```python
from itertools import combinations
from collections import Counter
count=Counter()
for row in df2["Grouped"]:
    row_list=row.split(',')
    #print(row_list)
    count.update(Counter(combinations(row_list,2)))

                 
for key,value in count.most_common():
                 print(key,value)
```

## 5th Insights: What product sold the most? Why it is sold the most?

```python
product_group=df.groupby("Product")
quantity_ordered=product_group.sum()["Quantity Ordered"]

products=[product for product,DF in product_group ]
#products
plt.bar(products,quantity_ordered)
plt.title("Representing the  Maximum Selling Quantity of Products")
plt.xlabel("Products")
plt.ylabel("Quantity")
plt.xticks(products,rotation="vertical",size=10)
plt.show()

```
